```
#GEOMETRIC SHAPES
from numpy import random
img3 = cv2.imread('/content/Salburg.jpg', 1)
img4 = cv2.imread('/content/Salburg.jpg', 1)

#rectangle
a = random.randint(img.shape[1]/2)
b = random.randint(img.shape[0]/2)
c = random.randint(img.shape[1]/2, img.shape[1])
d = random.randint(img.shape[0]/2, img.shape[0])
img3 = cv2.rectangle(img3, (a, b), (c, d), (242,179,177), 10)

#line
img3 = cv2.line(img3, (a, b), (c, d), (243,233,176), 10)
img3 = cv2.line(img3, (c, b), (a, d), (245,198,120), 10)

#circle
img3 = cv2.circle(img3, (int((a + c) / 2), int((b + d) / 2)), int(((c - a) ** 2 + (b - d) ** 2) ** 0.5 / 2), (152,218,200), 5)

#Ellipse
x = random.randint(img4.shape[1]/4, img4.shape[1]/2)
y = random.randint(img4.shape[0]/4, img4.shape[0]/2)
center_coordinates = (x, y)
axesLength = (150, 100)
angle = 40
startAngle = 0
endAngle = 360
color = (255, 0, 0)
thickness = -1
img4 = cv2.ellipse(img4,center_coordinates, axesLength, angle, startAngle, endAngle, color, thickness)
 
#polygon
pts = np.array([[100,100],[600,100],[400,500],[200,500]], np.int32)
pts = pts.reshape((pts.shape[0],1,pts.shape[1]))
img4 =cv2.polylines(img4,[pts],True,(0,255,0),5)

#text
font = cv2.FONT_HERSHEY_SIMPLEX
#show
plt.figure(figsize = (20, 20))
plt.subplot(1, 2, 1), plt.imshow(img3, interpolation = 'bicubic', cmap = 'gray')
plt.subplot(1, 2, 2), plt.imshow(img4, interpolation = 'bicubic')
print(img3.shape)
print(c)
```
