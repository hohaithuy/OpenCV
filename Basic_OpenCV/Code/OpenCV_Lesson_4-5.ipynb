```
#Libary will use
import matplotlib.pyplot as plt
import numpy as np
import cv2
img = cv2.imread('/content/drive/My Drive/content/statue.jpg', 1)
```

# **CHANGING COLORSPACES**

```
#2
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img)
plt.title('BGR')
plt.subplot(1, 3, 2), plt.imshow(img_RGB)
plt.title('RGB')
```

```
#3
img_GRAY = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_GRAY2BGR = cv2.cvtColor(img_GRAY, cv2.COLOR_GRAY2BGR)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img)
plt.title('BGR')
plt.subplot(1, 3, 2), plt.imshow(img_GRAY, cmap = 'gray')
plt.title('GRAY')
plt.subplot(1, 3, 3), plt.imshow(img_GRAY2BGR)
plt.title('GRAY->BGR')
```

```
#4
img_BGR2RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_RGB2GRAY = cv2.cvtColor(img_BGR2RGB, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img)
plt.title('BGR')
plt.subplot(1, 3, 2), plt.imshow(img_BGR2RGB)
plt.title('RGB')
plt.subplot(1, 3, 3), plt.imshow(img_RGB2GRAY, cmap = 'gray')
plt.title('GRAY')
```

```
#5
img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_HSV2BGR = cv2.cvtColor(img_HSV, cv2.COLOR_HSV2BGR)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img)
plt.title('BGR')
plt.subplot(1, 3, 2), plt.imshow(img_HSV)
plt.title('HSV')
plt.subplot(1, 3, 3), plt.imshow(img_HSV2BGR)
plt.title('HSV->BGR')
```

```
#6
img_HSV = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)
img_HSV2RGB = cv2.cvtColor(img_HSV, cv2.COLOR_HSV2RGB)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img_RGB)
plt.title('RGB')
plt.subplot(1, 3, 2), plt.imshow(img_HSV)
plt.title('HSV')
plt.subplot(1, 3, 3), plt.imshow(img_HSV2RGB)
plt.title('HSV->RGB')
```

```
#Object tracking
lower_blue = np.array([17, 20, 224]) 
upper_blue = np.array([130,255,255])
img_RGB2HSV = cv2.cvtColor(img_RGB, cv2.COLOR_RGB2HSV)
mask = cv2.inRange(img_RGB2HSV, lower_blue, upper_blue)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img_RGB)
plt.title('RGB')
plt.subplot(1, 3, 2), plt.imshow(img_RGB2HSV)
plt.title('RGB->HSV')
plt.subplot(1, 3, 3), plt.imshow(mask, cmap = 'gray')
plt.title('HSV->RGB')
```

# **SIMPLE THRESHOLDING**

```
#IMPORT IMAGE
img_threshold = cv2.imread('/content/drive/My Drive/content/threshold.jpg', 0)
img_gradient = cv2.imread('/content/drive/My Drive/content/gradient.tif', 0)
img_gradient_with_text = cv2.imread('/content/drive/My Drive/content/gradient_with_text.tif', 0)
img_noisy2 = cv2.imread('/content/drive/My Drive/content/noisy2.jpg', 0)
img_eight_pepper = cv2.imread('/content/drive/My Drive/content/eight_pepper.tif', 0)
```

```
#1
ret1, img_thresh_binary = cv2.threshold(img_threshold, 120, 255, cv2.THRESH_BINARY)
ret2, img_thresh_binary_inv = cv2.threshold(img_threshold, 120, 255, cv2.THRESH_BINARY_INV)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img_threshold, cmap = 'gray')
plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(img_thresh_binary, cmap = 'gray')
plt.title('THRESH_BINARY')
plt.subplot(1, 3, 3), plt.imshow(img_thresh_binary_inv, cmap = 'gray')
plt.title('THRESH_BINARY_INV')
```

```
#2
ret3, img_thresh_trunc = cv2.threshold(img_threshold, 127, 255, cv2.THRESH_TRUNC)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img_threshold, cmap = 'gray')
plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(img_thresh_trunc, cmap = 'gray')
plt.title('THRESH_TRUNC')
```

```
#3
ret1, img_thresh_tozero = cv2.threshold(img_gradient, 127, 255, cv2.THRESH_TOZERO)
ret2, img_thresh_tozero_inv = cv2.threshold(img_gradient, 127, 255, cv2.THRESH_TOZERO_INV)
plt.figure(figsize = (15, 15))
plt.subplot(1, 3, 1), plt.imshow(img_gradient, cmap = 'gray')
plt.title('Original')
plt.subplot(1, 3, 2), plt.imshow(img_thresh_tozero, cmap = 'gray')
plt.title('THRESH_TOZERO')
plt.subplot(1, 3, 3), plt.imshow(img_thresh_tozero_inv, cmap = 'gray')
plt.title('THRESH_TOZERO_INV')
```

```
#4
plt.figure(figsize = (20, 20))
ret1, img_gwt_thresh_binary = cv2.threshold(img_gradient_with_text, 127, 255, cv2.THRESH_BINARY)
img_adaptive_thresh_mean_c = cv2.adaptiveThreshold(img_gradient_with_text, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)
img_adaptive_thres_gaussian_c = cv2.adaptiveThreshold(img_gradient_with_text, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 2)
plt.subplot(1, 4, 1), plt.imshow(img_gradient_with_text, cmap = 'gray')
plt.title('Original')
plt.subplot(1, 4, 2), plt.imshow(img_gwt_thresh_binary, cmap = 'gray')
plt.title('THRESH_BINARY')
plt.subplot(1, 4, 3), plt.imshow(img_adaptive_thresh_mean_c, cmap = 'gray')
plt.title('ADAPTIVE_THRESH_MEAN_C')
plt.subplot(1, 4, 4), plt.imshow(img_adaptive_thres_gaussian_c, cmap = 'gray')
plt.title('ADAPTIVE_THRESH_GAUSSIAN_C')
```

```
#5
plt.figure(figsize = (15, 6))
plt.subplot(1, 2, 1), plt.imshow(img_eight_pepper, cmap = 'gray')
plt.title('Original')
plt.subplot(1, 2, 2), plt.hist(img_eight_pepper.ravel(), 256)
plt.title('Histogram')
```

```
#7
ret1, img_th1 = cv2.threshold(img_noisy2, 127, 255, cv2.THRESH_BINARY)
ret2, img_th2 = cv2.threshold(img_noisy2, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
blur = cv2.GaussianBlur(img_noisy2, (5, 5), 0)
ret3 , img_th3 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
images = [img_noisy2, 0, img_th1, img_noisy2, 0, img_th2, blur, 0, img_th3]
titles = ['Original Noisy Image', 'Histogram','Global Thresholding (v=127)','Original Noisy Image', '', "Otsu's Thresholding", 'Gaussian filtered Image', '', "Otsu's Thresholding"]
plt.figure(figsize = (15, 9))
for i in range (3):
  plt.subplot(3, 3, i * 3 + 1), plt.imshow(images[i * 3], 'gray')
  plt.title(titles[i * 3]), plt.xticks([]), plt.yticks([])
  plt.subplot(3,3,i*3+2),plt.hist(images[i*3].ravel(),256)
  plt.title(titles[i*3+1]),plt.xticks([]),plt.yticks([])
  plt.subplot(3,3,i*3+3),plt.imshow(images[i*3+2],'gray')
  plt.title(titles[i*3+2]),plt.xticks([]),plt.yticks([])
plt.show()
```
